```python
# Export to csv
df.to_csv("Atlantic_sample.csv")
```

```python
clean_df1 = df.loc[(df['date']==20210206)]
clean_df1
```

```python
df.to_csv("Atlantic_sample_06Feb2021.csv")
```

```python
clean_df1[['date', 'state','positive']]
```

```python
# Get rid of locations that are not part of the continental US
clean_df1.drop(clean_df1.loc[clean_df1['state']=='VI'].index, inplace=True)
clean_df1.drop(clean_df1.loc[clean_df1['state']=='GU'].index, inplace=True)
clean_df1.drop(clean_df1.loc[clean_df1['state']=='AS'].index, inplace=True)
clean_df1.drop(clean_df1.loc[clean_df1['state']=='PR'].index, inplace=True)
clean_df1.drop(clean_df1.loc[clean_df1['state']=='MP'].index, inplace=True)

clean_df1
```

```python
# resetindex values and making sure it doesn't add a new index column
clean_df2 = clean_df1.reset_index(drop=True, inplace=False)
clean_df2
```

```python
clean_df2
```

```python
clean_df3 = clean_df2[["date", "state", "positive"]]
clean_df3
```

```python
# Rename state column to match merge column
clean_df3.rename(columns = {'state':'STABBREV'}, inplace = True) 
```

```python
# write this as case data to join
clean_df3.to_csv("Cases_by_State_clean_06FEB21.csv")
```

```python
# Bring in Vaccine Data
VaccineData = "covid19_vaccinations_in_the_united_states_07Feb2021.csv"
```

```python
# Read and display the CSV with Pandas
Vax_df = pd.read_csv(VaccineData)
Vax_df.head()
```

```python
# Rename column
Vax_df.rename(columns = {'Unnamed: 1':'STABBREV'}, inplace = True) 
```

```python
Vax_df.head()
```

```python
# Pull only necessary columns
clean_Vax = Vax_df[["State", "STABBREV", "Tot_Deliv","Tot_Admin","First_Dose","Two_Doses"]]
clean_Vax.head()
```

```python
# merge vaccine and case data on state abbreviation
merge1 = clean_df3.merge(clean_Vax, how='outer', on='STABBREV')
```

```python
# Success - we see positive cases along with the vaccine data
merge1.head()
```

```python
# Write out merged dataset
merge1.to_csv("Merge1_cases_vax_06FEB21.csv")

# Note to avoid any issues later, use encoding="utf-8"
#merge1.to_csv("Merge1_cases_vax_06FEB21.csv", encoding="utf-8", index=False)
```
